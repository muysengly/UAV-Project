# Uniform Distribution

https://en.wikipedia.org/wiki/Continuous_uniform_distribution

```python
import numpy as np

num_gu = 5  # number of ground users

x_min = 0  # minimum x-axis [meter]
x_max = 100  # maximum x-axis [meter]

y_min = 0  # minimum y-axis [meter]
y_max = 100  # maximum y-axis [meter]

# generate random x-axis
gu_x = np.random.uniform(low=x_min, high=x_max, size=(num_gu,))

# generate random y-axis
gu_y = np.random.uniform(low=y_min, high=y_max, size=(num_gu,))

# generate z-axis
gu_z = np.zeros((num_gu,))

# combine xyz into array
gu_xyz = np.vstack([gu_x, gu_y, gu_z]).T

# show
gu_xyz
```

## 1. Plot the generated data

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))

for i in range(num_gu):
    plt.scatter(x=gu_xyz[i, 0], y=gu_xyz[i, 1], c="blue")
    plt.text(x=gu_xyz[i, 0], y=gu_xyz[i, 1], s=f"GU-{i}")


plt.xlim((x_min,x_max))
plt.ylim((y_min,y_max))

plt.xlabel("x-axis [m]")
plt.ylabel("y-axis [m]")

plt.grid()
```
