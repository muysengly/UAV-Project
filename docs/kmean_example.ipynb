# K-means clustering

link-1: https://en.wikipedia.org/wiki/K-means_clustering

link-2: https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html


```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

NUM_NODE = 100
DIMENSION = 2

node_xy = np.random.random((NUM_NODE, DIMENSION))
display(node_xy)
```

```python
NUM_CENTER = 20
kmeans = KMeans(
    n_clusters=NUM_CENTER,
    n_init="auto"
).fit(node_xy)
```

```python
labels = kmeans.labels_
labels
```

```python
centers = kmeans.cluster_centers_
centers
```

```python
# plot the results
plt.figure(figsize=(6, 6))

plt.scatter(
    node_xy[:, 0],
    node_xy[:, 1],
    label="Nodes", 
    c=labels,
)
plt.scatter(
    centers[:, 0],
    centers[:, 1],
    marker="x",
    c=range(20),
    label="Centers",
)

plt.xlim(0, 1)
plt.ylim(0, 1)

plt.xticks(np.linspace(0, 1, 11))
plt.yticks(np.linspace(0, 1, 11))

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend()
plt.grid()
plt.show()
```
