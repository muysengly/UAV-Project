# K-means clustering

link: https://en.wikipedia.org/wiki/K-means_clustering



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.vq import kmeans

NUM_NODE = 100
DIMENSION = 2

gu_xy = np.random.random((NUM_NODE, DIMENSION))
display(gu_xy)
```

```python
NUM_CENTER = 20

center = kmeans(obs=gu_xy, k_or_guess=NUM_CENTER)[0]
display(center)
```

```python
# plot the results
plt.figure(figsize=(6, 6))

plt.scatter(
    gu_xy[:, 0],
    gu_xy[:, 1],
    label="Nodes"
)
plt.scatter(
    center[:, 0],
    center[:, 1],
    marker="x",
    label="Centers"
)

plt.xlim(0, 1)
plt.ylim(0, 1)

plt.xticks(np.linspace(0, 1, 11))
plt.yticks(np.linspace(0, 1, 11))

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend()
plt.grid()
plt.show()
```
