```python
import time
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output

fig, ax = plt.subplots(figsize=(6, 6))

ax.set_ylim(0, 1)
ax.set_xlim(0, 10)


for i in range(10):

    ax.scatter(i, np.random.rand(), marker="x")

    display(fig)
    clear_output(wait=True)
    time.sleep(1)
```
