```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.vq import kmeans

gu_xy = np.random.random((100,2))
display(gu_xy)

center = kmeans(obs=gu_xy, k_or_guess=20)[0]
display(center)
```

```python
plt.scatter(gu_xy[:, 0], gu_xy[:, 1])
plt.scatter(center[:, 0], center[:, 1], marker="x")

plt.xlim(0, 1)
plt.ylim(0, 1)
plt.show()
```
